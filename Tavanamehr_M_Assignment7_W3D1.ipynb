#### 1. Use itertuples to group records by year, and display a summary of total points scored each year.

```python
def summary(ls):
  Count = len(ls)
  Mean = sum(ls)/len(ls)
  Min = min(ls)
  Max = max(ls)
  per25, Median, per75 = np.quantile(ls, q = np.arange(0.25, 1, 0.25))
  std = (sum([(i-Mean)**2 for i in ls])/Count)**0.5
  return Count, Mean, std, Min, per25, Median, per75, Max
```

```python
dataset_numberic = dataset.select_dtypes(include=['float64','int64'])
```

```python
dataset_numberic.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
```

```python
impKNN = KNNImputer(n_neighbors=5)
```

```python
newvalues = impKNN.fit_transform(dataset_numberic)
```

```python
dataset_numberic_col = dataset_numberic.columns
```
