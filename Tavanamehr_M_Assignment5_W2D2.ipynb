### 1. Write a Python program using filter function to check which strings in an input list are palindromes. A palindrome is a string that looks the same forwards and backwards. For example, given the list ("madam", "rocky", "adam", "anna"), your program should return ("madam", "anna"), a list containing only palindromes.

```python
def check_palindrome(data):
  for val in data:
    if data[0:]  != data[-1::-1]:
      return False
    return True
```

```python
data = ["madam", "rocky", "adam", "anna"]
list(filter(check_palindrome,data))
```

#### Create a series from the 'Illinois Surcharge' column. Show summary statistics for the series.

#### For the 'Illinois Surcharge' column, show only rows that contain amounts between the first and the third quartile.

```python
first_q = data_Illinois.quantile(q=0.25)
third_q = data_Illinois.quantile(q=0.75)
dataset[data_Illinois.map(lambda x: first_q <= x <= third_q)]
```

#### For the 'Illinois Surcharge' column, check if there are any zero entries. If yes, show their count.

```python
print('count zero entries in Illinois Surcharge column: {0}\n'.format(data_Illinois.map(lambda x: x == 0).sum()))
dataset[data_Illinois.map(lambda x: x == 0)]
```

#### For the 'Illinois Surcharge' column, show a mean for only counties whose population is above 50,000 people.

```python
data_Population = pd.Series(dataset.Population)
```

```python
data_Population = data_Population.apply(lambda x: x.replace(',','')).astype(int)
```

```python
print('mean of value in Illinois Surcharge when Population is more than 50,000 {0}',format(data_Illinois[data_Population.map(lambda x: x >= 50000)].mean()))
```
