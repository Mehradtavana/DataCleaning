```python
def checkmerge(dfleft, dfright, columnName):
    dfleft['inleft'] = 'Y'
    dfright['inright'] = 'Y'
    dfboth = pd.merge(dfleft[[columnName, 'inleft']], dfright[[columnName, 'inright']], on=[columnName], how="outer")
    dfboth.fillna('N', inplace=True)
    print(pd.crosstab(dfboth.inleft, dfboth.inright))
```

```python
checkmerge(NBA_dataset_1,NBA_dataset_2,'game_id')
```

```python
left_merge = pd.merge(NBA_dataset_1, NBA_dataset_2, on = ['game_id'], how='left')
right_merge = pd.merge(NBA_dataset_1, NBA_dataset_2, on = ['game_id'], how='right')
inner_merge = pd.merge(NBA_dataset_1, NBA_dataset_2, on = ['game_id'], how='inner')
outer_merge = pd.merge(NBA_dataset_1, NBA_dataset_2, on = ['game_id'], how='outer')
```

```python
print('left_join', str(left_merge.shape), '\n')
print('right_join', str(right_merge.shape), '\n')
print('inner_join', str(inner_merge.shape), '\n')
print('outer_join', str(outer_merge.shape), '\n')
```
