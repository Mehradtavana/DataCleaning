```python
dataset = pd.read_csv("ss.csv", encoding="ISO-8859-1")
```

```python
dataset.head()
dataset.dtypes
dataset.shape
dataset.info()
```

#### Read in all of the data and convert appropriate columns to a "Date" data type

```python
dataset_date = ['Ship Date','Order Date']
dataset[dataset_date] = dataset[dataset_date].astype('datetime64[ns]')
```

```python
dataset.dtypes
```

#### View the first few rows. Show the data type for all columns, as well as the number of rows and columns

```python
dataset.head()
dataset.dtypes
dataset.shape
```

#### Set the column ''Row ID" as the index column. Make sure to check if the column values are unique

```python
dataset['Row ID'].count()
dataset['Row ID'].nunique()
```

```python
dataset.set_index('Row ID',inplace=True)
```

```python
dataset.head()
```

#### Show the data types and non-null value counts for all columns

```python
dataset.info()
```

#### Convert all columns with object data type to category data type

```python
dataset.loc[:,dataset.dtypes == 'object'] = dataset.select_dtypes(['object']).apply(lambda x: x.astype('category'))
```

```python
dataset.dtypes
```

#### Use the bracket operator, loc, and iloc to select a few columns. Display your selections for each

```python
dataset.loc[:,dataset_date]
dataset.iloc[:,1:3]
```

#### Select all columns with numeric data types

```python
dataset_numeric = dataset.select_dtypes(include=["number"])
```

```python
dataset_numeric
```

#### Use slicing to start at the 1001st row and go to the 2000th row, skipping every other row

#### Select all the rows where 'Sales' exceed $100

```python
dataset_highsale = dataset.loc[dataset.Sales > 100]
```

```python
dataset_highsale.head()
dataset.shape
```

#### Select all rows dealing with 'office supplies' where 'Profit' is negative (i.e., there is loss)

```python
dataset_negativeprofit = dataset.loc[dataset.Profit < 0]
```

```python
dataset_negativeprofit.head()
dataset_negativeprofit.shape
```

#### Show the frequencies and percentages for columns 'Quantity' and 'Discount'

```python
# frequencies
dataset['Quantity'].value_counts()
dataset['Discount'].value_counts()
```

```python
# percentages
dataset['Quantity'].value_counts(normalize=True, sort=False)
dataset['Discount'].value_counts(normalize=True, sort=False)
```

#### Show the distribution of values for the 'Sales' and 'Discount' columns

```python
dataset_dist = ['Sales','Discount']
dataset[dataset_dist].describe()
```

```python
distribution_range = np.arange(0,1.1,0.1)
dataset[dataset_dist].quantile(distribution_range)
```

```python
distribution_range = np.arange(0,1.05,0.05)
dataset[dataset_dist].quantile(distribution_range)
```
