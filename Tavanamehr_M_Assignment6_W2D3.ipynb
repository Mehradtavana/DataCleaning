#### 3. Add a new column 'Kind of Earner' that will display a string, based on the value of the 'Other Salaries' column. The column shows 'Huge Earner' for salaries above 250,000, 'High Earner' for salaries over 150,000, 'Low Earner' for salaries under $50,000, and 'Medium Earner' for any other values. Show counts for the new column.

```python
dataset['Kind of Earner'] = np.where(dataset['Other Salaries']> 250000, 'Huge Earner', 
                                     np.where(dataset['Other Salaries'] > 150000, 'High Earner',
                                              np.where(dataset['Other Salaries'] < 50000, 'Low Earner', 'Medium Earner')))
```

```python
dataset['Kind of Earner'].value_counts()
```

#### 4. Create a column 'How Old' which will have values of either 'Brand New' for reports from year 2020, 'Old' for reports from up to year 2015, and 'Timely' for anything else. Use NumPY 'select' method to accomplish this task.

```python
dataset['Report Year'].value_counts().sort_index()
```

```python
condition = [dataset['Report Year'] == 2020, 
             (2015 < dataset['Report Year']) & (dataset['Report Year'] < 2020),
             True]
value = ['Brand New','Old','Timely']
dataset['How Old']=np.select(condition, value)
dataset['How Old'].astype('category')
```

```python
dataset['How Old'].value_counts()
```

#### 5. Create a column 'popularName' that displays 'Yes' if the first name of the individual for contacts is a popular first name, and 'No' is the first name of the contact is not popular. Popular first names to consider are: Michael, William, John, and Chris.

```python
Popular_name = ['Michael', 'William', 'John', 'Chris']
dataset['popularName'] = np.where(dataset.Contact.str.split(' ',expand=True)[0].isin(Popular_name),'Yes','No')
```

```python
dataset.sample(10)
```
